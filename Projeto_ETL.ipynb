```python
#função que substitui os espaços, remove acentos e caracteres epeciais

import re #com ela dá substituit os caracteres por _
from unidecode import unidecode #remove acentos

#criando função para renomear colunas
def renomear_colunas3(df2: pd.DataFrame) -> pd.DataFrame:
  # função para limpar os nomes das colunas, removendo espaço, acentos e caracteres especiais
  def limpar_colunas(col_name: str) -> str:
    col_name = unidecode(col_name) #retira os acentos
    col_name = re.sub(r'[^A-Za-z0-9_]+', '_', col_name) #subsitui qualquer caractere que não seja letra por _
    return col_name

  df2.columns = [limpar_colunas(col) for col in df2.columns]
  return df2
```

```python
# Função para criar o banco de dados
def criar_banco(conexao, database):
    try:
        cursor = conexao.cursor()
        cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
        conexao.commit()
        print(f"Banco de dados {database} criado ou já existente.")
    except Error as e:
        print(f"Erro ao criar banco de dados: {e}")

criar_banco(conexao, database)
```
